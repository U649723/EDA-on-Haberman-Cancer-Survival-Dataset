# EDA on Haberman Cancer Survival Dataset
<p>Exploratory Data Analysis (EDA) is the series of asking questions and applying statistics and visualization techniques to answer those questions and to uncover the hidden insights from the data. A case study on the cancer survival data set is done to explore the most common EDA techniques in this blog.</p>

```python
#importing the libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing the dataset
df=pd.read_csv("haberman.csv")
```

```python
np.shape(df)
```

```python
print (df.columns)
```

```python
#plotting the dataset using seaborn
df.plot(kind="scatter",x="Age",y="Year of Operations");
#kind type scatter
plt.grid()
#plot.grid will show the grid
plt.show()
#show function will show the plot result
```

```python
df.head()
```

```python
#this is the number who survived and not survived
df["Survival Status "].value_counts()
```

```python
#plotting the dataset using seaborn
df.plot(kind="scatter",x="Age",y="axillary nodes");
#kind type scatter
plt.grid()
#plot.grid will show the grid
plt.show()
#show function will show the plot result
```

```python
#adding the white grid style 
sns.set_style("whitegrid");
#adding datapoint colour and size
sns.FacetGrid(df, hue="Survival Status ",size=5)\
        .map(plt.scatter, "Age","axillary nodes")\
        .add_legend(); #legend fuction will describe class color
plt.show();
    
```

```python
#pairplot on Habeman dataset
plt.close()
sns.pairplot(df,hue="Survival Status ",vars=['Age', 'Year of Operations', 'axillary nodes'],size=4)
plt.show()
```

```python
def dens_plot(features,class_var):
    #adding the white grid style 
    sns.set_style(style="whitegrid")
    #adding datapoint colour and size
    sns.FacetGrid(data=df, hue=class_var,size=5)\
        .map(sns.distplot,features)\
        .add_legend()
```

```python
dens_plot("Age","Survival Status ")
plt.show()
```

# Observations from the above PDFs and CDFs :

```python
#Box Plot for the feature axillary nodes
sns.boxplot(x = 'Survival Status ', y ='axillary nodes', data = df)
plt.show()
#Box Plot for the feature Age
sns.boxplot(x = 'Survival Status ',y = 'Age', data = df)
plt.show()
#Box Plot for the feature Year of Operations
sns.boxplot(x = 'Survival Status ', y = 'Year of Operations', data = df)
plt.show()
```

```python
#Violin Plots for Axillary_Nodes
sns.violinplot(x = "Survival Status ", y = "axillary nodes", data = df, size = 8)
plt.show()

#Violin Plots for Operation_Year
sns.violinplot(x = "Survival Status ", y = "Year of Operations", data = df, size = 8)
plt.show()

#Violin Plots for Age
sns.violinplot(x = "Survival Status ", y = "Age", data = df, size = 8)
plt.show()
```

```python
sns.jointplot(data=df, x="Age",y="Year of Operations", kind="kde")
plt.ylabel("Cancer")
plt.show()
```
